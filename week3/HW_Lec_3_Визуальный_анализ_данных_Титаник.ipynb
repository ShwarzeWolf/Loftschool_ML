**Выкинем признак `Cabin`, а потом – все строки, где есть пропуски.**

```python
train_df = train_df.drop('Cabin', axis=1).dropna()
```

**Постройте попарные зависимости признаков `Age`, `Fare`, `Pclass`, `Sex`, `SibSp`, `Parch`, `Embarked` и `Survived`. (метод `scatter_matrix ` Pandas или `pairplot` Seaborn).**

```python
sex_map = {'male': 1, 'female': 2}
embarked_map = {'S': 0, 'C': 1, 'Q': 2}

train_df['Sex_values'] = train_df.apply(lambda x: sex_map.get(x['Sex']), axis=1)
train_df['Embarked_values'] = train_df.apply(lambda x: embarked_map.get(x['Embarked']), axis=1)

sns.pairplot(train_df[['Age', 'Fare', 'Pclass', 'Sex_values', 'SibSp', 'Parch', 'Embarked_values', 'Survived']])
```

**Как плата за билет (`Fare`) зависит от класса каюты (`Pclass`)? Постройте boxplot.**

```python
sns.boxplot(train_df['Pclass'], train_df['Fare'])
```

**Такой boxplot получается не очень красивым из-за выбросов.** 

**Cоздайте признак `Fare_no_out` (стоимости без выбросов), в котором исключаются стоимости, отличающиеся от средней по классу более чем на 2 стандартных отклонения. Важно: надо исключать выбросы именно в зависимости от класса каюты. Иначе исключаться будут только самые большие (1 класс) и малые (3 класс) стоимости.**

```python
train_df['Fare_no_out'] = train_df['Fare']
fare_pclass1 = train_df[train_df['Pclass'] == 1]['Fare']
fare_pclass2 = train_df[train_df['Pclass'] == 2]['Fare']
fare_pclass3 = train_df[train_df['Pclass'] == 3]['Fare']

fare_pclass1_no_out = fare_pclass1[abs(fare_pclass1 - fare_pclass1.mean()) < 2 * fare_pclass1.std()]
fare_pclass2_no_out = fare_pclass2[abs(fare_pclass2 - fare_pclass2.mean()) < 2 * fare_pclass2.std()]
fare_pclass3_no_out = fare_pclass3[abs(fare_pclass3 - fare_pclass3.mean()) < 2 * fare_pclass3.std()]

train_df['Fare_no_out'] = fare_pclass1_no_out.append(fare_pclass2_no_out) \
                                             .append(fare_pclass3_no_out)

sns.boxplot(train_df['Pclass'], train_df['Fare_no_out'])
```

**Каково соотношение погибших и выживших в зависимости от пола? Отобразите c помощью Seaborn.countplot c аргументом hue.**

```python
sns.countplot(train_df['Survived'], hue=train_df['Sex'])
```

**Каково соотношение погибших и выживших в зависимости от класса каюты? Отобразите c помощью Seaborn.countplot c аргументом hue.**

```python
sns.countplot(train_df['Survived'], hue=train_df['Pclass'])
```

**Как факт выживания зависит от возраста пассажира? Проверьте (графически) предположение, что молодые чаще выживали. Пусть, условно, молодые - младше 30 лет, пожилые – старше 60 лет.**

```python
def get_age_category(age):
    if age > 60:
        return 'alder'
    elif age < 30:
        return 'young'
    else:
        return 'middle'

train_df['Age_category'] = train_df.apply(lambda x: get_age_category(x['Age']), axis=1)
sns.countplot(train_df['Survived'], hue=train_df['Age_category'])
```
